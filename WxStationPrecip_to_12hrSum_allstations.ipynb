```python
import datetime
import numpy as np
import pandas as pd

import os
import os.path

stations = ["HurricaneRidge", "MtBakerSkiArea", "WashingtonPass","StevensPass","SnoqualmiePass","Paradise","WhitePass","MtHoodMeadows","Timberline"]
proj_dir   = '/Users/robertshahn/Documents/python-projects/Verification/Observed_Precip/'

for stn in stations:
    data_file = os.path.join(proj_dir, 'Precip_Obs_Extractor_'+ stn +'.csv')

    df = pd.read_csv(data_file)
    #print(df)
    pd.options.display.float_format = '{:,.2f}'.format
    #df = df.iloc[::-1]
    #df['datetime'] = pd.to_datetime(df['timecode'])
    df['datetime'] = pd.to_datetime(df['timecode'])
    df = df.set_index('datetime')
    df.drop(['timecode'], axis=1, inplace=True)

    #df.drop(['timecode'], axis=1, inplace=True)
    #df['12_hr sum'] = abs(round(df.rolling(12).sum(),2))
    df['12_hr sum'] = abs(round(df.rolling(12).sum(),2))
    print(df.iloc[300:400])
    #print(df)
    #decimals = pd.Series([2, 2], index=['datetime','12_hr_sum'])
    #df = df.round(decimals)
    df_12Z = df.at_time('04:00:00')
    df_12Z.drop('precipitation', axis=1, inplace=True)
    df_00Z = df.at_time('16:00:00')
    df_00Z.drop('precipitation', axis=1, inplace=True)
    outfile_12Z = os.path.join(proj_dir, stn + "_12Z.csv")
    outfile_00Z = os.path.join(proj_dir, stn + "_00Z.csv")
    df_12Z.to_csv(outfile_12Z)
    df_00Z.to_csv(outfile_00Z)

```
